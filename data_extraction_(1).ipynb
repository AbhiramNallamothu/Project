```python
custom_grp = data.groupby('Customization')
```

```python
new_df = None
for grp in custom_grp.groups:
    grp_data = custom_grp.get_group(grp)
    prompt_grp = grp_data.groupby("Optimal Prompt")
    for grp in prompt_grp.groups:
        prompt_grp_data = prompt_grp.get_group(grp)
        display(prompt_grp_data)
        display(prompt_grp_data[prompt_grp_data["Rating"] == max(prompt_grp_data["Rating"])].sample(1))
        if(new_df is not None):
            new_df.loc[len(new_df.index)] = prompt_grp_data[prompt_grp_data["Rating"] == max(prompt_grp_data["Rating"])].sample(1).values[0]
        else:
            new_df = prompt_grp_data[prompt_grp_data["Rating"] == max(prompt_grp_data["Rating"])].sample(1)
```
